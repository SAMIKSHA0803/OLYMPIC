```python
# IMPORTING THE IMPORTANAT LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# LOADING THE DATASET

athletes = pd.read_csv('D:/New folder/olympic/athlete_events.csv')
region = pd.read_csv('D:/New folder/olympic/noc_regions.csv')
```

```python
athletes.head(10)
```

```python
region.head(10)
```

```python
# JOINING THE DATAFRAMES

athletes_df = athletes.merge(region, how = 'left', on = 'NOC')
athletes_df.head(10)
```

```python
# TO KNOW THE TOTAL NUMBER OF ROWS AND COLUMNS

athletes_df.shape
```

```python
# TO MAKE THE COLUMN NAMES CONSISTENT

athletes_df.rename(columns = {'region':'Region', 'notes': 'Notes'}, inplace = True)
athletes_df.head(10)
```

```python
# TO GET INFORMATION ABOUT THE DATASET

athletes_df.info(10)
```

```python
# TO GET STATISTICAL INFERANCE

athletes_df.describe()
```

```python
# CHECKING FOR ANY NULL VALUES

nun_values = athletes_df.isna()
nan_columns = nun_values.any()
nan_columns
```

```python
# TOP 10 COUNTRIES PARTICIPATED IN OLYMPIC

top_10_countries = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)

top_10_countries
```

```python
# PLOT AS A GRAPH

plt.figure(figsize = (12, 6))
#plt
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y = top_10_countries, palette = 'Set2')
```

```python
# AGE DISTRIBUTION OF THE PARTICIPANTS

plt.figure(figsize = (12, 6))
plt.title("Age Distribution of the Athletes")
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'orange', edgecolor = 'white');
```

```python
# WINTER OLYMPIC SPORTS

winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
# SUMMER OLYMPIC SPORTS

summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
# MALE AND FEMALE PARTICIPANTS

gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# PIE PLOT FOR MALE AND FEMALE ATLETES

plt.figure(figsize = (12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f', startangle = 180, shadow = True);
```

```python
# TOTAL MEDALS ATHLETES WON

athletes_df.Medal.value_counts()
```

```python
# TOTAL NUMBER OF FEMALE ATHLETES IN EACH OLYMPIC

female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
# ANOTHER WAY

womenolympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]

womenolympics
```

```python
sns.set(style = "darkgrid")
plt.figure(figsize = (20, 10))
sns.countplot(x = 'Year', data = womenolympics, palette = "Spectral" )
plt.title('Women Participation')
```

```python
# TREND OF FEMALES PARTICIPATION IN OLYMPIC

part = womenolympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize = (20, 10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# GOLD MEDAL ATHLETES

goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals.head()
```

```python
# TAKE ONLY THE VALUES THAT ARE DIFFERENT FROM NAN

goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# GOLDMEDAL BEYOND 60 YEARS

goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] > 60]

sporting_event
```

```python
# PLOTTING THE SPORTING EVENT

plt.figure(figsize = (10, 5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold Medals for the Atheletes over 60 years Age')
```

```python
# Gold Medals from each country

goldMedals.Region.value_counts().reset_index(name = 'Medal').head(6)
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = "index", y = "Medal", data = totalGoldMedals, height = 5, kind = "bar", palette = "rocket")
g.despine(left = True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Numbers of Medals")
plt.title('Gold Medal per country')
```

```python
# RIO OLYMPIC

max_year = athletes_df.Year.max()
print(max_year)

team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

```python
# creating bar plot

sns.barplot(x = team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Countrywise Medals for the Year 2016');
```
